```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

```python
train = pd.read_csv("/Users/zoetsai/Downloads/pisa2009train.csv")
```

```python
test = pd.read_csv("/Users/zoetsai/Downloads/pisa2009test.csv")
```

```python
train.shape
```

```python
train.groupby("male").mean()
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
print(train.shape)
print(test.shape)
```

```python
for col in train.columns:
    print(train[col].value_counts())
```

```python
dummies = pd.get_dummies(train['raceeth'])
train = pd.concat([train, dummies], axis=1)
train = train.drop(['raceeth', 'White'], axis=1)
```

```python
import re
train.columns = [re.sub("[ /-]", "", c) for c in train.columns]
```

```python
train.describe()
```

```python
train.columns
```

```python
all_columns = "+".join(train.columns - ["readingScore"])
model = sm.ols(formula="readingScore ~ " + all_columns, data=train).fit()
```

```python
model.summary()
```

```python
np.sqrt(np.mean(model.resid ** 2))
```

```python
dummies = pd.get_dummies(test['raceeth'])
test = pd.concat([test, dummies], axis=1)
test = test.drop(['raceeth', 'White'], axis=1)
test.columns = [re.sub("[ /-]", "", c) for c in test.columns]
```

```python
test.describe()
```

```python
preds = model.predict(test)
print(preds.min())
print(preds.max())
print(preds.max() - preds.min())
```

```python
resids = preds - test['readingScore']
sse = (resids**2).sum()
rmse = np.sqrt((resids**2).mean())
print(sse)
print(rmse)
```

```python
preds_baseline = np.array([train['readingScore'].mean()] * test.shape[0])
```

```python
resids_baseline = preds_baseline - test['readingScore']
sse_baseline = (resids_baseline**2).sum()
rmse_baseline = np.sqrt((resids_baseline**2).mean())
print(sse_baseline)
print(rmse_baseline)
```

```python
sst = sse_baseline
test_r2 = 1.0 - (sse/sst)
print(test_r2)
```
