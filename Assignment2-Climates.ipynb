```python
import pandas as pd
import statsmodels.formula.api as sm
import numpy as np
import itertools
```

```python
climate.columns = ['Year', 'Month', 'MEI', 'CO2', 'CH4', 'N2O', 'CFC11', 'CFC12', 'TSI', 'Aerosols', 'Temp']
```

```python
train = climate[climate['Year'] <= 2006].copy()
test = climate[climate['Year'] > 2006].copy()
```

```python
# AIC steps
AICs = {}
target = train['Temp']
predictorcols = ['Year', 'Month', 'MEI', 'CO2', 'CH4', 'N2O', 'CFC11', 'CFC12', 'TSI', 'Aerosols']
for k in range(1,len(predictorcols)+1):
    for variables in itertools.combinations(predictorcols, k):
        predictors = train[list(variables)]
        predictors['Intercept'] = 1
        res = sm.OLS(target, predictors).fit()
        AICs[variables] = 2*(k+1) - 2*res.llf
pd.Series(AICs).idxmin()
```

```python
variables = pd.Series(AICs).idxmin()
```

```python
model3 = sm.OLS(target, train[list(variables)]).fit()
```

```python
model3.summary()
```
