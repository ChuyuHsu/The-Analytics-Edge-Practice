```python
import statsmodels.formula.api as sm
```

```python
columns = songs.columns.tolist()
novars = set(["Top10", "year", "songtitle", "artistname", "songID", "artistID"])
columns = [c for c in columns if not c in novars]
print(columns)
```

```python
st = "+".join(columns)
SongsLog1 = sm.logit(formula="Top10 ~ " + st, data=train).fit()
```

```python
SongsLog1.summary()
```
